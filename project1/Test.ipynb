```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
import os
```

```python
from paths import CAR_IMAGES_DIR  # YOU NEED TO GO TO THE paths.py FILE AND PUT YOUR FOLDER OF IMAGES
```

```python
INDIVIDUAL_IMG_PATHS = [CAR_IMAGES_DIR / image_name for image_name in os.listdir(CAR_IMAGES_DIR)]
```

```python
def binary_image(image_path):
    image = cv.imread(str(image_path))
    image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    binary__image = cv.threshold(image_gray, 140, 255, 0)
    return binary__image[1]
```

```python
fig, axis = plt.subplots(3, 3, figsize = (10, 10))
for i in range(3):
    for j in range(3):
        image = binary_image(random.choice(INDIVIDUAL_IMG_PATHS))
        axis[i, j].imshow(image, cmap=plt.cm.gray)

plt.show()
```
